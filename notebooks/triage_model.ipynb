```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
import pickle
```

```python
data = {
    "symptoms": [
        "chest pain", "breathing problem", "high fever",
        "headache", "cough", "heart pain", "vomiting"
    ],
    "severity": [
        "CRITICAL", "URGENT", "NORMAL",
        "LOW", "LOW", "CRITICAL", "NORMAL"
    ]
}

import pandas as pd
df = pd.DataFrame(data)
df
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df["symptoms"])
```

```python
y = df["severity"]

model = LogisticRegression()
model.fit(X, y)

print("Model trained")
```

```python
test = vectorizer.transform(["chest pain"])
print(model.predict(test))
```

```python
import pickle

pickle.dump(model, open("../app/models/triage_model.pkl", "wb"))
pickle.dump(vectorizer, open("../app/models/vectorizer.pkl", "wb"))

print("Model saved")
```
